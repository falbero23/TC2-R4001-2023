#### Verificación polos de la transferencia sobre la circunferencia de radio unitario

```python
#%% Librerias %%#
from scipy.signal import TransferFunction
from pytc2.sistemas_lineales import pzmap, bodePlot
#%% Transferencia %%#
tf = TransferFunction( [1], [1, 2, 2, 1] ) # 1 / (S³ + 2S² + 2S + 1)
bodePlot(tf, fig_id=1, filter_description = 'Vo/Vi')
pzmap(tf, fig_id=2, filter_description = 'Vo/Vi')
```
