### Ejercicio 4

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Apr 18 15:10:16 2023

@author: fede23
"""

#%% Ejercicio n°4 - Trabajo Semanal 1 %%#

from scipy.signal import TransferFunction
from pytc2.sistemas_lineales import pzmap, bodePlot

# Transferencia normalizada: T($) = ($ - R2/R1) / ($ - 1)
wp = 1 # Frecuencia del polo
wo = [0.5, 1, 1.5] # 3 posibles frecuencias para el cero, dependiendo de la relación de componentes R2/R1

for i in range(len(wo)):
    
    tf_n = TransferFunction( [1, -1*(wo[i])], [1, 1] )
    
    bodePlot( tf_n, fig_id=1, filter_description = 'R2/R1 = {:3.3f}'.format(wo[i]) )
    
    pzmap( tf_n, fig_id=2, filter_description = 'R2/R1 = {:3.3f}'.format(wo[i]) )
```

### Ejercicio 5

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Apr 18 16:01:11 2023

@author: fede23
"""

#%% Ejercicio n°5 - Trabajo Semanal 1 %%#

from scipy.signal import TransferFunction
from pytc2.sistemas_lineales import pzmap, bodePlot

# Transferencia sin normalizar (ej 1): T(S) = ( S - R2/(R1*R3*C1) ) / ( S + 1/(R3*C1) )
# T(S) = ( S - (R2/R1)*Wo ) / ( S + Wp )
# Como R2/R1=1 ---> Wo = Wp = w = 1/(R3*C1)

R3 = 1e3
C1 = 1e-6
w = 1 / (R3*C1)

tf = TransferFunction([1, -1*w], [1, w])
bodePlot(tf, fig_id=1, filter_description = 'R2/R1 = 1')
pzmap(tf, fig_id=2, filter_description = 'R2/R1 = 1')
```

### Simulación circuital en LTspice

#### Análisis de módulo y fase de la trasferencia $T_{(s)}$ :

\begin{equation}
T(s) = \frac{V_{2 (s)}}{V_{1 (s)}} = \frac{ S - \frac{R _2}{R _1 R _3 C _1} }{ S + \frac{1}{R _3 C _1} } 
\end{equation}

<img src="grafico_TS1_LTSpice.png" title="Grafico TS1 LTSpice" >

***Aclaraciones***: Se corroboran los resultados obtenidos a través de la simulación numérica para el caso de $\frac{R_2}{R_1}=1$ (Filtro pasa todo), con una ganancia de aproximadamente 0dB y una fase que va desde $\pi$ en $w=0$ hasta 0 en $w \to  \infty$, pasando por $\frac{\pi}{2}$ en $w=w_o=\frac{1}{R_3C_1}$.
Esta simulación se ha llevado a cabo con un modelo real del OPAMP, ya que en alta frecuencia (arriba de los 10KHz) la transferencia empieza a caer debido al producto ganancia - ancho de banda (GBW).
Debido a que se utiliza un modelo real del OPAMP, se obtiene que $|T_{(s)}|=-173.8 udB$, lo cual es virtualmente una ganancia de 0 dB para toda frecuencia $w$, comprobando así la característica de funcionamiento del *filtro pasa todo*.
